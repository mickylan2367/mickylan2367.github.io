```python
import requests

# HORIZONSシステムのベースURL
base_url = "https://ssd.jpl.nasa.gov/horizons_batch.cgi"
api_key = "RnOwoYyB08ELa6g7RSxGcGuJfFaaMXlwXXo9AIpa"

# POSTリクエストのパラメータ
# 以下は太陽の位置情報を取得する例です。他の天体を取得する場合、目標物体のIDを変更してください。
payload = {
    "batch": "1",        # バッチモードを有効にする
    "make_ephem": "YES",  # エフェメリス（位置情報）を作成
    "table_type": "VECTORS",
    "reference_plane": "ECLIPTIC",
    "center": "10",       # 中心のID（太陽）
    "start_time": "now",  # 現在の日時から計算
    "stop_time": "now+1d",  # 現在の日時から1日後までの情報を取得
    "step_size": "1d",     # 1日ごとのステップ
    "quantities": "1,9",   # 1: X軸（Ecliptic longitude）、9: Y軸（Ecliptic latitude）
    "CSV_FORMAT": "YES",   # CSVフォーマットでデータ取得
    "COMMAND": "'10' 'lon@10' 'lat@10'",  # クエリ
}

# HORIZONSシステムにPOSTリクエストを送信
response = requests.post(base_url, data=payload)

# レスポンスを表示
if response.status_code == 200:
    print(response.text)
else:
    print(f"エラー：{response.status_code} - {response.text}")
```
